```python
import sys
sys.path.append("../../")
DATA_PATH = "../../data/"
SAVE_PATH = "../../plots"

import matplotlib.pyplot as plt
import numpy as np
import py_common.utility as utility
```

Single Channel

```python
data = np.loadtxt(f'{DATA_PATH}/single_chan/wave_function.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("position [bohr]")
ax.set_ylabel("wave function")

ax.plot(data[:, 0], data[:, 1] / np.max(np.abs(data[:, 1])))
ax.plot(data[:, 0], data[:, 3] / (0.03 * np.max(np.abs(data[:, 3]))))
ax.plot(data[:, 0], data[:, 2] / (np.max(np.abs(data[:, 2]))))
ax.set_ylim(-1.1, 1.1);
```

```python
data = np.loadtxt(f'{DATA_PATH}/single_chan/propagation_distance.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("position [bohr]")
ax.set_ylabel("scattering length [bohr]")

ax.plot(data[:, 0], data[:, 1])
ax.plot(data[:, 0], data[:, 2]);
```

```python
data = np.loadtxt(f'{DATA_PATH}/single_chan/mass_scaling.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("mass scaling factor")
ax.set_ylabel("scattering length [bohr]")

utility.plot_scattering(ax, data[:, 0], data[:, 1], data[:, 2])
ax.set_ylim(-1000, 1000);
```

```python
data = np.loadtxt(f'{DATA_PATH}/single_chan/node_count.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("energy [GHz]")
ax.set_ylabel("bound difference [a.u.]")
ax.set_ylim(-10, 11)

ax.plot(data[:, 0], data[:, 1]);
```

Two channel

```python
data = np.loadtxt(f'{DATA_PATH}/two_chan/wave_function.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("position [bohr]")
ax.set_ylabel("wave function")

normalization_nodes = np.max(np.abs(data[:, 3]))

normalization = np.max(np.abs(data[:, 1]))
ax.plot(data[:, 0], data[:, 1] * normalization_nodes / normalization)

normalization = np.max(np.abs(data[:, 2]))
ax.plot(data[:, 0], data[:, 2] * normalization_nodes / normalization)

ax.plot(data[:, 0], data[:, 3])
# ax.set_ylim(-3, 30)
# ax.set_xlim(7.5, 40);
```

```python
data = np.loadtxt(f'{DATA_PATH}/two_chan/mass_scaling.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("mass scaling factor")
ax.set_ylabel("scattering length [bohr]")

utility.plot_scattering(ax, data[:, 0], data[:, 1], data[:, 2])
ax.set_ylim(-1000, 1000);
```

```python
ns = [0, 1, 3, -1, -2, -5]

fig, ax = utility.plot()
ax.set_xlabel("position [bohr]")
ax.set_ylabel("wave function")
ax.set_ylim(-1.1, 1.1)

for n in ns:
    data = np.loadtxt(f'{DATA_PATH}/single_chan/bound_wave_{n}.dat', delimiter='\t', skiprows=1)

    normalization = np.max(np.abs(data[:, 1]))
    ax.plot(data[:, 0], data[:, 1] / normalization)
```
