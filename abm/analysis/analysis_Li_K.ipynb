```python
import sys
sys.path.append("../../")
DATA_PATH = "../../data/"
SAVE_PATH = "../../plots"

import matplotlib.pyplot as plt
import numpy as np
import py_common.utility as utility
```

```python
data = np.loadtxt(f'{DATA_PATH}/Li_K/hifi_Li6.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("Magnetic Field [G]")
ax.set_ylabel("Energy [GHz]")

ax.plot(data[:, 0], data[:, 1::], color="blue")
plt.savefig(f"{SAVE_PATH}/hifi_Li6.pdf", format='pdf')
```

```python
data = np.loadtxt(f'{DATA_PATH}/Li_K/hifi_K40.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("Magnetic Field [G]")
ax.set_ylabel("Energy [GHz]")

ax.plot(data[:, 0], data[:, 1::], color="blue")
plt.savefig(f"{SAVE_PATH}/hifi_K40.pdf", format='pdf')
```

```python
data = np.loadtxt(f'{DATA_PATH}/Li_K/abm.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("Magnetic Field [G]")
ax.set_ylabel("Energy [GHz]")

ax.plot(data[:, 0], data[:, 1::], color="blue", linewidth = 1)
ax.set_ylim(-2.0, -0.6)
ax.set_xlim(0.0, 400.0)

hifi_Li  = np.loadtxt(f'{DATA_PATH}/Li_K/hifi_Li6.dat', delimiter='\t', skiprows=1)
hifi_K = np.loadtxt(f'{DATA_PATH}/Li_K/hifi_K40.dat', delimiter='\t', skiprows=1)

ax.plot(hifi_Li[:, 0], hifi_K[:, 2] + hifi_Li[:, 1], color="red", linewidth = 1, linestyle = "--")

plt.savefig(f"{SAVE_PATH}/abm_Li6_K40_mf_-3.pdf", format='pdf')
```
