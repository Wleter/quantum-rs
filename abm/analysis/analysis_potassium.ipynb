```python
import sys
sys.path.append("../../")
DATA_PATH = "../../data/"
SAVE_PATH = "../../plots"

import matplotlib.pyplot as plt
import numpy as np
import py_common.utility as utility
```

# 39-41 calculation

```python
data = np.loadtxt(f'{DATA_PATH}/potassium/hifi_39_41.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("Magnetic Field [G]")
ax.set_ylabel("Energy [GHz]")

ax.plot(data[:, 0], data[:, 1::], color="blue");
```

```python
data = np.loadtxt(f'{DATA_PATH}/potassium/abm_39_41.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("Magnetic Field [G]")
ax.set_ylabel("Energy [GHz]")

ax.plot(data[:, 0], data[:, 1::], color="blue", linewidth = 1)
data_hifi = np.loadtxt(f'{DATA_PATH}/potassium/hifi_39_41.dat', delimiter='\t', skiprows=1)
ax.plot(data_hifi[:, 0], data_hifi[:, 1], color="red", linewidth = 1, linestyle = "--")
fig.savefig(f"{SAVE_PATH}/potassium_39_41.pdf", bbox_inches='tight')
```
